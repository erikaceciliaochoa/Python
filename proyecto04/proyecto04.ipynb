## Loop/ciclo "for"

```python
numeros = [1,2,3,4,5]
for numero in numeros: # por cada nro en esta lista hacer esto
	print(numero)
```

## Loop/ciclo "while"

```python
contador = 5
while contador > 0: #mientras coontador mayor a cero haga lo siguiente 
	print(contador)
	contador = contador - 1
```
